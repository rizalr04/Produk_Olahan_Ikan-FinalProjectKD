```python
import pandas as pd
import numpy as np

poi=pd.read_csv('/content/produkolahanikan.csv ')
poi
```

```python
s=poi[['JenisOlahan','Tahun','Volume']]
s
```

```python
import plotly.express as px
fig = px.scatter(s,                       #DataFrame yang akan di plot
                 x="Tahun",          #Data untuk sumbu X
                 y="Volume",         #Data untuk sumbu Y
                 color="JenisOlahan",          #Memberikan Warna (Opsional)
                       #Memberikan Ukuran (Opsional)
)

#Menambahkan Keterangan
fig.update_layout(
    #Menambahkan dan mengatur Judul Grafis
    title='Data Jenis Olahan Ikan Berdasarkan Tahun', 
    title_font_color="blue",
    title_font_size=30,    
    #Menambahkan dan mengatur label sumbu x            
    xaxis=dict(
        title='Tahun',
        titlefont_size=16,
        tickfont_size=14),
    #Menambahkan dan mengatur label sumbu y
    yaxis=dict(
        title='Volume',
        titlefont_size=16,
        tickfont_size=14))
#Menampilkan Figure
fig.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

df = pd.read_csv (r'/content/produksibudidayanasional.csv')
print (df)
```

```python
fig = px.bar(df, x="NamaIkan", y="Nilai")
fig.show()
```

Berdasarkan grafik diatas dapat disimpulkan bahwa komoditas tertinggi dalam Produksi Budidaya Nasional berasal dari komoditas total.

```python
fig = px.bar(df, x="Tahun", y="Nilai")
fig.show()
```

Berdasarkan grafik diatas dapat disimpulkan bahwa setiap tahunnya nilai dari Produksi Budidaya Nasional mengalami peningkatan setiap tahunnya.

```python
fig = px.scatter(df, x="NamaIkan", y="Nilai", color='NamaProvinsi')
fig.show()
```

Berdasarkan grafik sebaran diatas dapat disimpulkan bahwa komoditas total memiliki persebaran yang banyak di setiap provinsinya.

```python
df = pd.read_csv (r'/content/produksiperlakuanikannasional.csv')
print (df)
```

```python
fig = px.bar(df, x="Wilayah", y="Volume", color='JenisPerlakuan')
fig.show()
```

Berdasarkan grafik diatas dapat disimpulkan bahwa komoditas tertinggi dalam Produksi Perlakuan Nasional berasal dari wilayah Sumatera.

```python
fig = px.bar(df, x="Tahun", y="Volume", color='JenisPerlakuan')
fig.show()
```

Berdasarkan grafik diatas dapat disimpulkan bahwa Produksi Perlakuan Nasional mengalami beberapa kali penurunanan, yaitu pada tahun 2004 dan 2009.

```python
fig = px.scatter(df, x="JenisPerlakuan", y="Volume", color='JenisPerairan')
fig.show()
```

Berdasarkan grafik persebaran diatas dapat disimpulkan bahwa Produksi Perlakuan Nasional banyak dilakukan pada jenis perairan laut diandingkan jenis perairan umum.
